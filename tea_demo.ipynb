```python
import pathlib

import requests

import tea
```

```python
# Load data

data_path = pathlib.Path("./datasets/UScrime.csv")
data_path.parent.mkdir(parents = True, exist_ok = True)

data_url = "https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/MASS/UScrime.csv"
data_path.write_text(requests.get(data_url).text)

tea.data(data_path)
```

```python
# Declare and annotate the variables of interest
variables = [
    {
        'name' : 'So',
        'data type' : 'nominal',
        'categories' : ['0', '1']
    },
    {
        'name' : 'Prob',
        'data type' : 'ratio',
        'range' : [0,1]
    }
]

tea.define_variables(variables)
```

```python
assumptions = {
    'groups normally distributed': [['So', 'Prob']],
    'Type I (False Positive) Error Rate': 0.05,
}

tea.assume(assumptions)
```

```python
experimental_design = {
                        'study type': 'observational study',
                        'contributor variables': 'So',
                        'outcome variables': 'Prob',
                    }
tea.define_study_design(experimental_design)
```

```python
tea.hypothesize(['So', 'Prob'], ['So:1 > 0'])
```
